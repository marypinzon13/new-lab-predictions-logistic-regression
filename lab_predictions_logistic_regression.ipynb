```python
query = '''SELECT *
FROM sakila.film as fi
INNER JOIN sakila.inventory as inv
ON fi.film_id = inv.film_id
INNER JOIN sakila.rental as re
ON inv.inventory_id = re.inventory_id;'''

data = pd.read_sql_query(query, engine)
data.head(60)
```

```python
query = '''SELECT fi.title as title, SUM(pay.amount) as amount
FROM sakila.film as fi
INNER JOIN sakila.inventory as inv
ON fi.film_id = inv.film_id
INNER JOIN sakila.rental as re
ON inv.inventory_id = re.inventory_id
INNER JOIN sakila.payment as pay
ON re.rental_id = pay.rental_id
GROUP BY fi.title ;'''

data = pd.read_sql_query(query, engine)
data.head()
```

```python
query = '''SELECT fi.title, COUNT(re.rental_id) as june 
FROM sakila.film as fi
INNER JOIN sakila.inventory as inv
ON fi.film_id = inv.film_id
INNER JOIN sakila.rental as re
ON inv.inventory_id = re.inventory_id
INNER JOIN sakila.payment as pay
ON re.rental_id = pay.rental_id
WHERE rental_date BETWEEN '2005-06-01' AND '2005-06-30'
GROUP BY fi.title
ORDER BY title;'''
data = pd.read_sql_query(query, engine)
data.head()
```
